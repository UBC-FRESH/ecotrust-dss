```python
import geopandas as gpd
import fiona

# Path to the geodatabase file
# gdb_path = '/media/data/project/sghotb/testtttt/destination_folder/tsa01.gdb'
gdb_path = '/media/data/project/sghotb/ecotrust_dss/data/gdb_files/tsa01.gdb'

# List layers in the geodatabase using Fiona
layers = fiona.listlayers(gdb_path)
# print("Available layers:", layers)

# Choose a specific layer to convert
layer_name = layers[0]  # Replace with the desired layer name if known

# Read the selected layer
gdf = gpd.read_file(gdb_path, layer=layer_name)

# Path to save the shapefile
shp_path = "./data/shp_files/output_file.shp"

# Save to a shapefile
gdf.to_file(shp_path, driver="ESRI Shapefile")

print(f"Shapefile saved to: {shp_path}")
```

```python


# gdb_path = '/media/data/project/sghotb/testtttt/destination_folder/tsa01.gdb'

# Open the geodatabase
with fiona.open(gdb_path, 'r') as src:
    # Get the list of layers
    layers = list(src.keys())
    # print(f"Layers in the geodatabase: {layers}")

    # Check if there are layers
    if layers:
        # Choose the first layer
        layer_name = layers[0]
        print(f"Reading attributes from layer: {layer_name}")

        # Use `iter` to correctly access features
        feature_iter = iter(src)
        try:
            # Get the first feature
            first_feature = next(feature_iter)
            print(f"First feature: {first_feature}")  # Inspect the first feature

            # Collect all attributes
            attributes = []
            for feature in feature_iter:
                attributes.append(feature['properties'])
            
            # Convert the attributes list into a pandas DataFrame
            import pandas as pd
            df = pd.DataFrame(attributes)
            print(df.head())

        except StopIteration:
            print("No features found in the layer.")

    else:
        print("No layers found in the geodatabase.")
```

```python
stands = df
```

```python
stands['primary_forest'] = (stands['Age2015'] > (60 - 10)).astype(int)
stands
```
